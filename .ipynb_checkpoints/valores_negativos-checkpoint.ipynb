```python

def cuotas_a_probabilidad(cuotas):
    temp=[]
    a_priori=1
    for i in range(3): 
        temp.append([1/cuotas[i][0],1/cuotas[i][1],cuotas[i][2]]) 
        a_priori*=cuotas[i][0]
    return a_priori,temp

def computar_true_p(cuotas):
    
    a_priori,temp=cuotas_a_probabilidad(cuotas)
    
    print("%"*50)
    print(temp)
    print("%"*50)
    
    prob_dos_primeros=1
    ganaron_primero=0
    
    if(temp[-1][2][0]>temp[-1][2][1]): #Gana
        prob_dos_primeros=temp[-2][1]*temp[-1][1] #P(G2|G1)P(G1)
        ganaron_primero=1
    else: #Pierde o empata
        prob_dos_primeros=temp[-2][0]-temp[-2][1]*(1-temp[-1][1]) #P(G2)-P(G2|~G1)P(~G1)
    
    print("IMPRIMIR",prob_dos_primeros,temp[-3][0])
        
    if(prob_dos_primeros<0):
        print("NEGATIVO EN PRIMA")
        return a_priori,10000

    prob_final=1
    
    if(temp[-2][2][0]>temp[-2][2][1] and ganaron_primero==1): #Ganaron los dos partidos previos
        prob_final=temp[-3][1]*prob_dos_primeros
    else:
        prob_final=temp[-3][0]-temp[-3][1]*(1-prob_dos_primeros)

    if(prob_final<0):
        print("NEGATIVO")    
        return a_priori,10000
    
    return a_priori,1/prob_final


def explotar(casa,cuotas_pais,df_prob_reales):
    
    equipos_infra=[]
    equipos_sobre=[]
    negativos=[]
        
    for i in cuotas_pais:
        print("*"*50)
        print(i)
        print(cuotas_pais[i][casa])
                 
        #a_priori,a_posteriori=computar_true_p(cuotas_pais[i][casa])
        a_priori,a_posteriori=computar_true_p(df_prob_reales[i])

        
        print("Priori vs posteriori",a_priori,a_posteriori)
        print("*"*50)

        if(a_posteriori==10000): negativos.append(i)
        
        if(a_posteriori<a_priori): equipos_infra.append((i,a_priori,a_posteriori))
        else: equipos_sobre.append((i,a_priori,a_posteriori))
        
        
    #Sortear por profit o loss
    monto=1
    dinero=0

    for i in equipos_infra:
        
        prob_real=1/i[2]
        diff=prob_real*(i[1]-1)*monto-(1-prob_real)*monto
         
        dinero=diff+dinero
        #if(i[0] not in paises_rank): dinero=diff+dinero
    
    print("DINERO +",dinero)
    
    sort_perder=[]
    
    for i in equipos_sobre:

        prob_real=1/i[2]
        diff=prob_real*(i[1]-1)*monto-(1-prob_real)*monto
        sort_perder.append((i[0],diff))
        dinero=diff+dinero

        #if(i[0] not in paises_rank): dinero=diff+dinero
    
    print("Negativos")
    for i in negativos:
        print(i)
    
    print("Infravalorados")
    
    for i in equipos_infra:
        print(i)
    
    print()
    print("Sobrevalorados")
    
    for i in equipos_sobre:
        print(i)
    
    print("Dinero despues",dinero)
    print("-"*50)
    
    return dinero
```

```python
matches = [
    "qatar-ecuador",
    "england-iran",
    "senegal-netherlands",
    "usa-wales",
    "argentina-saudi-arabia",
    "denmark-tunisia",
    "mexico-poland",
    "france-australia",
    "morocco-croatia",
    "germany-japan",
    "spain-costa-rica",
    "belgium-canada",
    "switzerland-cameroon",
    "uruguay-south-korea",
    "portugal-ghana",
    "brazil-serbia"
]

def comparar_1(odds):
    
    print("Gana 1",np.mean(odds[0]),np.mean(odds[3]))
    
    if(np.mean(odds[0])<np.mean(odds[3])):
        print(odds[0])
        print(odds[3])
    
    print("Gana 2",np.mean(odds[2]),np.mean(odds[5]))

    if(np.mean(odds[2])<np.mean(odds[5])):
        print(odds[2])
        print(odds[5])
        
def comparar(odds):
    
    for i in range(len(odds)):
        profit_antes=0
        profit_despues=0

        for j in range(3): profit_antes+=1/odds[i][j]
        for j in range(3): profit_despues+=1/odds[i][j+3]

        print(profit_antes,profit_despues)

def normalize_decimal_odds(odds):
    # Convert decimal odds to probabilities
    probabilities = 1 / np.array(odds)
    
    total_probability = np.sum(probabilities)
    normalized_probabilities = probabilities / total_probability
    
    # Convert the normalized probabilities back to decimal odds
    
    #print(1 / normalized_probabilities)
        
    normalized_odds = 1 / normalized_probabilities
    
    return normalized_odds

def sumar_consenso(arr_original,antes,despues):
    for i in range(3): arr_original[i].append(antes[i])
    for i in range(3): arr_original[i+3].append(despues[i])
        
def prob_consenso(arr_original):
    
    #print("CONSENSO")
    
    devolver_cuotas=[]
    
    for i in range(6):

        sd=0
        mean=np.sum(arr_original[i])/len(arr_original[i])
        
        for j in arr_original[i]: sd+=(j-mean)*(j-mean)
        
        sd=np.sqrt(sd/len(arr_original[i]))
        
        media_nueva=0
        num=0
        for j in arr_original[i]:
            if(abs(j-mean)/sd<=1): 
                num+=1
                media_nueva+=j
                
        devolver_cuotas.append(media_nueva/num)
    
    devolver_cuotas[0:3]=normalize_decimal_odds(devolver_cuotas[0:3])
    devolver_cuotas[3:6]=normalize_decimal_odds(devolver_cuotas[3:6])
    
    return devolver_cuotas
    
def analisis(year):
    global matches
    df_general=pd.read_csv(f"./data_{year}/todo.csv")
    df_casas={}
    df_prob_reales=defaultdict(lambda:[])
    cuotas_pais=defaultdict(lambda:defaultdict(lambda:[]))
    casas_validas=defaultdict(lambda:1)
    sumar=0
    
    for i in casas:
        try: df_casas[i]=pd.read_csv(f"./data_{year}/{i}.csv")
        except: casas_validas[i]=0

    for index,rows in df_general.iterrows():
    
        if(rows["tipo"]=="Groups"):

            pais_a=rows["pais_a"]
            pais_b=rows["pais_b"]
            encuentro=rows["encuentro"]
            res_a=rows["res_a"]
            res_b=rows["res_b"]
            
            arr_consenso=[[],[],[],[],[],[]]
            num=0
            
            for i in casas:

                try:
                
                    odds_casa=df_casas[i][df_casas[i]["encuentro"]==encuentro]
                    casas_validas[i]=1*casas_validas[i]
                    
                    #normalizar_cuotas
                    #meterlas en true_odds
                    cuotas_a_priori=[odds_casa["inicial_0"].values[0],odds_casa["inicial_x"].values[0],odds_casa["inicial_1"].values[0]]
                    cuotas_a_posteriori=[odds_casa["final_1"].values[0],odds_casa["final_x"].values[0],odds_casa["final_2"].values[0]]
                    
                    sumar_consenso(arr_consenso,cuotas_a_priori,cuotas_a_posteriori)
                    
                    #cuotas_a_priori=normalize_decimal_odds(cuotas_a_priori)
                    #cuotas_a_posteriori=normalize_decimal_odds(cuotas_a_posteriori)
                    
                    cuotas_pais[pais_a][i].append([cuotas_a_priori[0],cuotas_a_posteriori[0],(res_a,res_b)])
                    cuotas_pais[pais_b][i].append([cuotas_a_priori[2],cuotas_a_posteriori[2],(res_b,res_a)])
                                
                                                                                   
                except Exception as e:
                    print(f"La casa {i}, no saco cuota para el partido {encuentro}")
                    casas_validas[i]=0*casas_validas[i]
                    
            if(encuentro in matches):
                print()
                print(encuentro)
                comparar(arr_consenso)
                print()
            nuevas_cuotas=prob_consenso(arr_consenso)
            df_prob_reales[pais_a].append([nuevas_cuotas[0],nuevas_cuotas[3],(res_a,res_b)])
            df_prob_reales[pais_b].append([nuevas_cuotas[2],nuevas_cuotas[5],(res_b,res_a)])

    avg=0
    validas=0
    
    for i in casas_validas:
        if(casas_validas[i]==1 and i=="bet365"):
            validas+=1
            print("-"*50)
            print(i)
            avg+=explotar(i,cuotas_pais,df_prob_reales)
            print("-"*50)
            break

    print(avg/len(casas_validas))
```
