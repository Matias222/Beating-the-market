```python
def get_std(arr):
    media=np.mean(arr)
    sd=0
    for i in arr:
        sd+=(i-media)*(i-media)/len(arr)
    sd=np.sqrt(sd)
    return sd

def extraer_data(year):
    
    df_general=pd.read_csv(f"./data_{year}/todo.csv")
    df_casas={}
    df_encuentros={}
    
    casas_validas=defaultdict(lambda:1)
    
    for i in casas:
        try: 
            casas_validas[i]=1
            df_casas[i]=pd.read_csv(f"./data_{year}/{i}.csv")
        except: casas_validas[i]=0

    for index,rows in df_general.iterrows():
    
        if(rows["tipo"]=="Groups"):

            pais_a=rows["pais_a"]
            pais_b=rows["pais_b"]
            encuentro=rows["encuentro"]
            
            inicial_0=[]
            inicial_x=[]
            inicial_1=[]
            final_0=[]
            final_x=[]
            final_1=[]
            casas_temp=[]
            
            for i in casas:

                odds_casa=df_casas[i][df_casas[i]["encuentro"]==encuentro]
                
                if(len(odds_casa)==0):
                    print(f"La casa {i}, no saco cuota para el partido {encuentro}")
                    casas_validas[i]=0
                else:
                
                    odds_casa=df_casas[i][df_casas[i]["encuentro"]==encuentro]
                    
                    casas_temp.append(i)
                    
                    inicial_0.append(odds_casa["inicial_0"].values[0])
                    inicial_x.append(odds_casa["inicial_x"].values[0])
                    inicial_1.append(odds_casa["inicial_1"].values[0])
                    
                    final_0.append(odds_casa["final_1"].values[0])
                    final_x.append(odds_casa["final_x"].values[0])
                    final_1.append(odds_casa["final_2"].values[0])
            
            df_encuentros[encuentro]=((inicial_0,inicial_x,inicial_1,final_0,final_x,final_1,casas_temp))
                      
    return df_encuentros
```

```python
df_encuentros=extraer_data(2022)
```

```python
def min_max(temp,casa):
    
    minimo=1000
    maximo=-1000
    minimo_casa=""
    maximo_casa=""
    
    for i in range(len(casa)):
        if(minimo>temp[i]):
            minimo=temp[i]
            minimo_casa=casa[i]
        if(maximo<temp[i]):
            maximo=temp[i]
            maximo_casa=casa[i]
    
    return minimo,minimo_casa,maximo,maximo_casa
            
def analizar(df_encuentros):
    #Como saco la varianza de un partido, el promedio de las 3? (En teoria estan relacionadas)
    
    top_inicial=[]
    top_final=[]
    
    for i in df_encuentros:
        #print("-"*50)
        #print(i)        
        std_inicial_0=get_std(df_encuentros[i][0])
        std_final_0=get_std(df_encuentros[i][-4])
        top_inicial.append((std_inicial_0,i))
        top_final.append((std_final_0,i))
        #minimo,minimo_casa,maximo,maximo_casa=min_max(df_encuentros[i][0],df_encuentros[i][-1])
        #print(minimo,minimo_casa)
        #print(maximo,maximo_casa)
        #print(std_inicial_0,std_final_0)
        #print("-"*50)

    top_inicial.sort(key=lambda x:-x[0])
    top_final.sort(key=lambda x:-x[0])
    
    for i in range(5):
        print("-"*50)
        print(top_inicial[i][1])
        media=np.mean(df_encuentros[top_inicial[i][1]][0])
        a,b,c,d=min_max(df_encuentros[top_inicial[i][1]][0],df_encuentros[top_inicial[i][1]][-1])
        print(a,(np.abs(a-media))/top_inicial[i][0],b)
        print(c,(np.abs(c-media))/top_inicial[i][0],d)
        print("-"*50)
        

analizar(df_encuentros)
```
